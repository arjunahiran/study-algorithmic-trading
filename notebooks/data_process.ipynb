```python
import pandas as pd

# read data
ticker = "MSFT"
df = pd.read_excel(f"../data/{ticker}.xlsx", parse_dates=["Date"], index_col=0)
df = df.loc["2000-01-01":, :].copy()
df
```

```python
# add tomorrow change %
df["tomorrow_change"] = df["Close"].pct_change(-1) * 100 * (-1)
df
```

```python
# remove NaN
df = df.dropna().copy()
df
```

```python
# add tomorrow's direction
df["tomorrow_direction"] = df["tomorrow_change"].map(lambda x: 1 if x > 0 else -1)
df
```

```python
# save processed data
df.to_excel(f"../data/{ticker}_processed.xlsx")
```
