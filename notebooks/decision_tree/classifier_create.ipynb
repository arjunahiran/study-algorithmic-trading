```python
import pandas as pd

ticker = "MSFT"

# load data
df = pd.read_excel(
    f"../../data/{ticker}_processed.xlsx", parse_dates=["Date"], index_col=0
)
df
```

```python
# X, y data
y = df["tomorrow_direction"]
y
```

```python
X = df.drop(columns=["tomorrow_change", "tomorrow_direction"])
X
```

```python
from sklearn.tree import DecisionTreeClassifier

# create a decision tree classifier model
model_dt = DecisionTreeClassifier(max_depth=15)
model_dt.fit(X, y)
```

```python
y_pred = model_dt.predict(X)
y_pred
```

```python
# compare
df_comp = df[["tomorrow_direction"]].copy()
df_comp["prediction"] = y_pred
df_comp
```

```python
# score
model_dt.score(X, y)
```

```python
import pickle

with open(f"models/{ticker}.pkl", "wb") as f:
    pickle.dump(model_dt, f)
```
